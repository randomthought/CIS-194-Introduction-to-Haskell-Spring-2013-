```haskell
{-# OPTIONS_GHC -Wall #-}
module LogAnalysis where
import Log
```

```haskell
parseMessage :: String -> LogMessage
parseMessage ('I':(x:xs))     = LogMessage Info x xs 
parseMessage ('W':(x:xs))     = LogMessage Warning x xs
parseMessage ('E':(e:(x:xs))) = LogMessage (Error e) x xs
parseMessage s                = Unknown s
```
